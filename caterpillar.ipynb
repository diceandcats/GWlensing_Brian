```python
# code for making a caterpillar plot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/home/dices/Research/GWlensing_Brian/src_pos_for_dist_with_z_de+mcmc.csv')

dz=abs(data['z']-data['localized_z'])
#print(dz)
# add the dz column to the dataframe
data['dz'] = dz.round(12)
# calculate dA by dxdy
dx = abs(data['x'] - data['localized_x'])
dy = abs(data['y'] - data['localized_y'])
data['dx'] = dx.round(12)
data['dy'] = dy.round(12)
data['dS'] = np.sqrt(dx**2 + dy**2).round(12)
# Print the dataframe with 12 decimal places
print(data)
# save the dataframe in original file
#data.to_csv('/home/dices/Research/GWlensing_Brian/src_pos_for_dist_with_z_de+mcmc.csv', index=False)
```

```python
# sample data
rng = np.random.default_rng(seed=42)

# event labels 1‒15
events = np.arange(1, 16)

# ----------------------------- x [arcsec]
x_med      = rng.normal( 0.00, 0.40, 15).round(3)      # medians
x_sigma_lo    = rng.uniform(0.10, 0.25, 15).round(3)
x_sigma_hi    = rng.uniform(0.15, 0.35, 15).round(3)
x_err      = np.vstack([x_sigma_lo, x_sigma_hi])             # ±σ for Matplotlib

# ----------------------------- y [arcsec]
y_med      = rng.normal(-0.10, 0.35, 15).round(3)
y_sigma_lo    = rng.uniform(0.10, 0.30, 15).round(3)
y_sigma_hi    = rng.uniform(0.10, 0.25, 15).round(3)
y_err      = np.vstack([y_sigma_lo, y_sigma_hi])             # symmetric ±σ

# ----------------------------- z_s (source redshift)
z_med      = rng.normal(2.50, 0.30, 15).round(3)
z_err_lo   = rng.uniform(0.15, 0.30, 15).round(3)      # 16th percentile → median
z_err_hi   = rng.uniform(0.20, 0.45, 15).round(3)      # median → 84th percentile
z_err      = np.vstack([z_err_lo, z_err_hi])           # asymmetric errors
```

```python

plt.rcParams.update({'axes.labelsize': 15, 'ytick.labelsize': 12})

# fig, ax = plt.subplots(1, 3, figsize=(10, 6), sharey=True)
# fig.suptitle('Caterpillar plots for 15 lensed‑GW events', fontsize=16)

# def _caterpillar(ax, centre, err, xlabel):
#     ax.errorbar(centre, events, xerr=err, fmt='o', capsize=4,
#                 elinewidth=1.3, mec='k', mew=0.8, ms=5)
#     ax.set_xlabel(xlabel)
#     ax.axvline(0, color='grey', lw=.8, ls='--')     # optional reference
#     ax.invert_yaxis()                               # event 1 at the top
#     ax.set_yticks(events)

# _caterpillar(ax[0], x_med, x_err, r'$x$ [arcsec]')
# _caterpillar(ax[1], y_med, y_err, r'$y$ [arcsec]')
# _caterpillar(ax[2], z_med, z_err, r'$z_s$')

# plt.tight_layout()
# plt.show()

fig, ax = plt.subplots(1, 3, figsize=(9, 6), sharey=True)
fig.suptitle('Uncertainty of Localized Parameters for Simulated Events', fontsize=15)

def _width_only(ax, err, label, first=False):
    ax.errorbar(  # event index on y‑axis, dummy x=0
        x=np.zeros(err.shape[1]), y=events,
        xerr=err, fmt='o',                        # hide marker
        ecolor='tab:blue', elinewidth=1.3, capsize=4, mec='k', mew=0.8, ms=5)
    ax.axvline(0, color='grey', lw=.8)            # zero baseline
    ax.set_xlabel(label);  ax.invert_yaxis()                       # event 1 at the top
    ax.tick_params(labelsize=14)
    ax.set_yticks(events)                     # event labels on y‑axis
    if first == True: 
        ax.set_ylabel('Events')       # only the first plot gets ticks

_width_only(ax[0], x_err, '±σₓ [arcsec]', first=True)
_width_only(ax[1], y_err, '±σᵧ [arcsec]')
_width_only(ax[2], z_err, '$\pm σ_z$')

plt.tight_layout();  plt.show()
```
